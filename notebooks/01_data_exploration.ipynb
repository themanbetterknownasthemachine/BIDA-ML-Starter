# ML TEMPLATE — EXPLORATIVE DATENANALYSE (EDA)

---

**Dieses Template hilft bei der ersten Analyse eines neuen Datensatzes.**
Ziel: Den Datensatz verstehen, bevor Modelle trainiert werden.

| Schritt | Inhalt |
|---------|--------|
| 1 | Imports & Daten laden |
| 2 | Erster Überblick |
| 3 | Fehlende Werte |
| 4 | Verteilungen |
| 5 | Korrelationen & Zusammenhänge |
| 6 | Zeitliche Muster (falls Zeitreihe) |
| 7 | Ausreisser |
| 8 | Fazit & nächste Schritte |

---

## 1. Imports & Daten laden
- Libraries importieren (pandas, numpy, matplotlib, seaborn)
- Config laden
- Daten aus Snowflake laden mit `load_query()` oder `load_table()`
- Oder: CSV/Excel laden für erste Tests

---

## 2. Erster Überblick
- `.shape` — Wie viele Zeilen und Spalten?
- `.head()` — Erste 5 Zeilen anschauen
- `.info()` — Datentypen und Nicht-Null Counts
- `.describe()` — Statistische Zusammenfassung (numerische Spalten)
- `.dtypes` — Welche Spalten sind numerisch, welche kategorisch?
- `.nunique()` — Wie viele eindeutige Werte pro Spalte?

---

## 3. Fehlende Werte
- Anzahl und Prozent fehlender Werte pro Spalte
- Visualisierung: Heatmap oder Barplot der fehlenden Werte
- Muster erkennen: Fehlen Werte zufällig oder systematisch?
- Entscheidung: Zeilen löschen, imputieren (Median, Mean, Mode) oder belassen?

---

## 4. Verteilungen
- Histogramme für numerische Spalten
- Boxplots für Ausreisser-Erkennung
- Barplots für kategorische Spalten (Häufigkeiten)
- Verteilung der Zielvariable:
  - Bei Regression: Schiefe (Skewness), Q-Q Plot
  - Bei Klassifikation: Klassenbalance prüfen
  - Bei Zeitreihen: Saisonalität, Trend

---

## 5. Korrelationen & Zusammenhänge
- Korrelationsmatrix (Heatmap) für numerische Spalten
- Top-Korrelationen mit der Zielvariable
- Scatterplots für die stärksten Zusammenhänge
- Bei kategorischen Variablen: Groupby + Aggregation

---

## 6. Zeitliche Muster (falls Zeitreihe)
- Zeitreihen-Plot (y über die Zeit)
- Wochentags-Muster (Boxplot pro Wochentag)
- Monats-Durchschnitte (Barplot pro Monat)
- Jahresvergleich (gleicher Monat, verschiedene Jahre)
- Lücken in der Zeitreihe prüfen (fehlende Tage/Monate?)

---

## 7. Ausreisser
- IQR-Methode oder Z-Score
- Visualisierung: Boxplots, Scatterplots
- Sonderereignisse identifizieren (Feiertage, Systemausfälle, Sonderaktionen)
- Entscheidung: Entfernen, ersetzen oder belassen?

---

## 8. Fazit & nächste Schritte
- Zusammenfassung: Was wurde entdeckt?
- Datenqualität bewerten (gut, mässig, schlecht)
- Welche Features sind vielversprechend?
- Welche Probleme müssen vor dem Modellieren gelöst werden?
- Empfehlung: Welches Template-Notebook als nächstes?
  - Zeitreihe → `02_forecasting.ipynb`
  - Klassifikation → `03_classification.ipynb`
  - Regression → `04_regression.ipynb`
