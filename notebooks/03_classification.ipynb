# ML TEMPLATE — CLASSIFICATION

---

**Dieses Template enthält alle wichtigen Schritte für ein ML-Klassifikations-Projekt:**

1. Imports & Daten laden
2. Explorative Datenanalyse (EDA)
3. Feature Engineering & Vorbereitung
4. Pipeline mit Preprocessing
5. Cross-Validation
6. Hyperparameter-Tuning (GridSearchCV)
7. Finales Training
8. Evaluation
9. Feature Importance
10. Modell speichern & laden
11. Zusammenfassung

---

## 1. Imports & Daten laden
- Alle benötigten Libraries importieren
- Random Seed setzen für Reproduzierbarkeit
- Warnungen konfigurieren
- Plotting-Style definieren
- Daten aus Snowflake laden mit `load_query()` oder `load_table()`
- Erste Übersicht mit `.head()`, `.info()`, `.shape`
- Datentypen prüfen
- Zielvariable definieren

---

## 2. Explorative Datenanalyse (EDA)
- Statistische Zusammenfassung (`.describe()`)
- Fehlende Werte analysieren
- Verteilung der Zielvariable (Klassenbalance prüfen)
- Bei starker Imbalance: Strategien überlegen (SMOTE, Class Weights, Undersampling)
- Korrelationsmatrix
- Visualisierungen (Histogramme, Boxplots, Pairplots)

---

## 3. Feature Engineering & Vorbereitung
- Fehlende Werte behandeln
- Neue Features erstellen
- Features transformieren (Log, Sqrt, etc.)
- Kategorische Variablen identifizieren
- Numerische und kategorische Spalten trennen
- Outlier behandeln
- Features selektieren
- Train/Test Split (Stratified bei Klassifikation!)

---

## 4. Pipeline mit Preprocessing
- ColumnTransformer für unterschiedliche Feature-Typen
- Numerisch: Imputation, Scaling
- Kategorisch: Imputation, Encoding (OneHot, Target, Ordinal)
- Pipeline zusammenbauen (Preprocessor + Classifier)
- DummyClassifier als absolute Baseline (most_frequent)
- Baseline Accuracy und F1 messen

---

## 5. Cross-Validation
- K-Fold mit Stratification (typisch k=5 oder k=10)
- Mehrere Metriken evaluieren:
  - Accuracy
  - F1-Score (weighted)
  - ROC-AUC
- Varianz der Ergebnisse prüfen
- Train-Score vs. Test-Score vergleichen (Overfitting?)

---

## 6. Hyperparameter-Tuning (GridSearchCV)
- Suchraum definieren (n_estimators, learning_rate, max_depth, num_leaves)
- Methode wählen:
  - GridSearchCV (exhaustiv, bei kleinem Suchraum)
  - RandomizedSearchCV (schneller, bei grossem Suchraum)
  - Optuna (intelligent, bayesische Optimierung)
- Beste Parameter speichern
- Besten CV-Score dokumentieren

---

## 7. Finales Training
- Bestes Modell mit optimalen Parametern
- Training auf gesamtem Trainingsset
- Oder: Refit auf Train+Validation (GridSearchCV mit refit=True)

---

## 8. Evaluation
- Vorhersagen auf Testset
- Classification Report (Precision, Recall, F1 pro Klasse)
- Confusion Matrix visualisieren
- ROC-Kurve und AUC berechnen
- Precision-Recall-Kurve
- Performance vs. Baseline vergleichen

---

## 9. Feature Importance
- Built-in Feature Importance (LightGBM/XGBoost)
- Permutation Importance
- SHAP Values für Erklärbarkeit
- Top Features visualisieren mit `plot_feature_importance()`
- Business-Interpretation: Machen die Top-Features Sinn?

---

## 10. Modell speichern & laden
- Modell mit joblib/pickle speichern über `save_model()`
- Gesamte Pipeline inklusive Preprocessing speichern (nicht nur den Classifier!)
- Laden und Testen: Predictions müssen identisch sein
- Versionierung dokumentieren (v1, v2, ...)

---

## 11. Zusammenfassung
- Ergebnisse dokumentieren
- Performance vs. Baseline: Wie viel besser?
- Limitationen benennen
- Verbesserungsvorschläge
- Deployment-Überlegungen
- Experiment loggen mit `log_experiment()`
