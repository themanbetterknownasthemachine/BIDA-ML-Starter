---

## 1. Imports & Konfiguration
- Alle benötigten Libraries importieren
- Random Seed setzen für Reproduzierbarkeit
- Warnungen konfigurieren
- Plotting-Style definieren
- Config laden aus `configs/config.yaml`

---

## 2. Daten laden
- Daten aus Snowflake laden mit `load_query()` oder `load_table()`
- Erste Übersicht mit `.head()`, `.info()`, `.shape`
- Datentypen prüfen
- Zielvariable definieren

---

## 3. Explorative Datenanalyse (EDA)
- Statistische Zusammenfassung (`.describe()`)
- Fehlende Werte analysieren
- Verteilung der Zielvariable (Schiefe, Normalität)
- Korrelationsmatrix
- Top Korrelationen mit der Zielvariable
- Visualisierungen (Histogramme, Boxplots, Scatterplots)
- Q-Q Plot für Normalitätsprüfung

---

## 4. Feature Engineering
- Fehlende Werte behandeln
- Neue Features erstellen
- Features transformieren (Log, Sqrt, etc.)
- Kategorische Variablen identifizieren
- Numerische und kategorische Spalten trennen
- Outlier behandeln

---

## 5. Train/Test Split
- Zufälliger Split für Regression (80/20 oder 70/30)
- Random State für Reproduzierbarkeit
- Verteilung der Zielvariable in Train und Test vergleichen

---

## 6. Outlier-Analyse Zielvariable
- Verteilung der Zielvariable prüfen
- Schiefe (Skewness) berechnen und interpretieren
- IQR-Methode für Outlier-Erkennung
- Transformation bei Schiefe testen:
  - Log-Transformation (nur bei positiven Werten)
  - Box-Cox
  - Yeo-Johnson (funktioniert auch mit negativen Werten)
- Entscheiden: Transformation anwenden oder nicht?
- Extreme Werte identifizieren und dokumentieren

---

## 7. Pipeline mit Preprocessing
- ColumnTransformer für unterschiedliche Feature-Typen
- Numerisch: Imputation (Median), Scaling (StandardScaler)
- Kategorisch: Imputation (Most Frequent), Encoding (OneHot, Target, Ordinal)
- Pipeline zusammenbauen (Preprocessor + Regressor)
- Alternativ: `build_preprocessor()` aus `src/preprocessing.py` nutzen

---

## 8. Baseline Model
- Einfaches Modell ohne Tuning trainieren
- DummyRegressor als absolute Baseline (Mean oder Median)
- Erste Performance-Messung: MAE, RMSE, R²
- Benchmark für alle weiteren Verbesserungen

---

## 9. Cross-Validation
- K-Fold (typisch k=5 oder k=10)
- Mehrere Metriken evaluieren:
  - RMSE (Root Mean Squared Error)
  - MAE (Mean Absolute Error)
  - R² (Coefficient of Determination)
- Varianz der Ergebnisse prüfen
- Train-Score vs. Test-Score: Overfitting?

---

## 10. Hyperparameter-Tuning
- Suchraum definieren
- Methode wählen:
  - GridSearchCV (exhaustiv, bei kleinem Suchraum)
  - RandomizedSearchCV (schneller, bei grossem Suchraum)
  - Optuna (intelligent, bayesische Optimierung)
- Typische LightGBM-Parameter:
  - n_estimators, learning_rate, max_depth
  - num_leaves, min_child_samples
  - reg_alpha, reg_lambda
- Beste Parameter speichern

---

## 11. Finales Training
- Bestes Modell mit optimalen Parametern
- Training auf gesamtem Trainingsset
- Oder: Refit auf Train+Validation

---

## 12. Evaluation
- Vorhersagen auf Testset
- Metriken berechnen mit `compute_regression_metrics()`:
  - RMSE, MAE, MAPE
  - R², Adjusted R²
- Actual vs. Predicted Plot
- Performance vs. Baseline vergleichen

---

## 13. Residuen-Analyse
- Residuen berechnen (Actual - Predicted)
- Residuen-Plots mit `plot_residuals()`:
  - Residuen vs. Fitted Values (Muster erkennen)
  - Histogramm der Residuen (Normalverteilung?)
  - Q-Q Plot (Normalität)
- Heteroskedastizität prüfen (Varianz der Fehler gleichmässig?)
- Muster in Fehlern erkennen (systematische Abweichungen?)
- Residuen-Statistiken: Mean (≈ 0?), Std, Schiefe

---

## 14. Feature Importance
- Built-in Feature Importance (LightGBM/XGBoost)
- Permutation Importance
- SHAP Values für Erklärbarkeit
- Top Features visualisieren mit `plot_feature_importance()`
- Business-Interpretation: Machen die Top-Features Sinn?

---

## 15. Modell speichern & laden
- Modell mit joblib/pickle speichern über `save_model()`
- Gesamte Pipeline inklusive Preprocessing speichern
- Laden und Testen: Predictions müssen identisch sein
- Versionierung dokumentieren (v1, v2, ...)

---

## 16. Zusammenfassung & nächste Schritte
- Ergebnisse dokumentieren
- Performance vs. Baseline: Wie viel besser?
- Limitationen benennen
- Verbesserungsvorschläge
- Deployment-Überlegungen
- Experiment loggen mit `log_experiment()`
